# 1. PLOT BASIC LINE PLOT

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read The data from stock prices
stock_df = pd.read_csv('stock_data.csv')
stock_df.head()
```

```python
stock_df.plot(x='Date' ,y='FB',label='Facebook Stock Price',figsize =(12,5), linewidth=3)
plt.ylabel('Prcie [$]')
plt.xlabel('Dates')
plt.title('Facebook Stock Price with Dates')
plt.legend(loc ='upper right')
plt.grid()
plt.show()
```

```python
#Plot similar kind of graph for NFLX
#Change the line color to red and increase the line width
stock_df.plot(x = 'Date', y = 'NFLX', label = 'Netflix Stock Price', figsize = (15, 10), linewidth = 7, color = 'r');
plt.ylabel('Price')
plt.title('My first plotting exercise!')
plt.legend(loc = "upper left")
plt.grid()
```

# 3. PLOT PIE CHART

```python
# Use matplotlib to plot a pie chart 
values = [20, 55, 5, 17, 3]
colors = ['g', 'r', 'y', 'b', 'm']
labels = ['AAPL', 'GOOG','T','TSLA','AMZN']
explode = [0,0,0,0,0.2]
plt.figure(figsize = (10, 10))
plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('STOCK PORTFOLIO')
plt.show()
```

```python
#Plot the pie chart for the same stocks assuming equal allocation
#Explode Amazon and Google slices
values = [20, 20, 20, 20, 20]
colors = ['g', 'r', 'y', 'b', 'm']
explode = [0, 0.2, 0, 0, 0.2]
labels = ['AAPL', 'GOOG', 'T', 'TSLA  ', 'AMZN']
plt.figure(figsize = (10, 10))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('STOCK PORTFOLIO')
plt.show()
```

# 4. PLOT HISTOGRAMS

```python
# A histogram represents data using bars of various heights. 
# Each bar groups numbers into specific ranges. 
# Taller bars show that more data falls within that specific range.
mu = daily_return_df['FB'].mean()
sigma = daily_return_df['FB'].std()
num_bins = 40
plt.hist(daily_return_df['FB'],num_bins,facecolor ='red')
plt.grid()
plt.title('Histogram: mu ='+ str(mu) +', sigma =' + str(sigma))
plt.show()
```

```python
num_bins = 30
plt.figure(figsize = (10,7))
plt.hist(daily_return_df['TWTR'], num_bins, facecolor = 'blue');
plt.grid()
```

# 8. SEABRON SCATTERPLOT & COUNTPLOT

```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import seaborn as sns # Statistical data visualization
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
# Check out the head of the dataframe
df_cancer.head()
```

```python
# Check out the tail of the dataframe
df_cancer.tail()
```

```python
# Plot scatter plot between mean area and mean smoothness
plt.figure(figsize = (10,10))
sns.scatterplot(x ='mean area', y ='mean smoothness', hue='target', data=df_cancer)
plt.show()
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1
plt.figure(figsize = (10,10))
sns.countplot(x= df_cancer['target'],label='count')
plt.show()
```

```python
#Plot the scatterplot between the mean radius and mean area. Comment on the plot
sns.scatterplot(x = 'mean radius', y = 'mean area', hue = 'target', data = df_cancer);
```

# 9. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS

```python
# Plot the pairplot
sns.pairplot(df_cancer, hue='target', vars=['mean radius','mean texture','mean area','mean perimeter','mean smoothness'])
plt.show()
```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize = (30, 30)) 
sns.heatmap(df_cancer.corr(),annot=True)
plt.show()
```

```python
# plot the histplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 
sns.histplot(df_cancer['mean radius'], bins=25, color='b')
plt.show()
```

```python
#Plot two separate histplot for each target class #0 and target class #1
class_0_df = df_cancer[ df_cancer['target'] == 0];
class_1_df = df_cancer[ df_cancer['target'] == 1];

class_0_df

class_1_df

# Plot the histplot for both classes
plt.figure(figsize=(10, 7));

sns.histplot(class_0_df['mean radius'], bins = 25, color = 'blue');
sns.histplot(class_1_df['mean radius'], bins = 25, color = 'red');
plt.grid();
plt.show()
```
